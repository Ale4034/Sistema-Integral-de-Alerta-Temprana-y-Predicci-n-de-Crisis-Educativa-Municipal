# 🧹 Módulo 3. Limpieza y transformación de datos

## Objetivo

En este módulo se realizan todas las transformaciones necesarias para convertir los tres conjuntos de datos originales en versiones consistentes y listas para su integración. Las decisiones de limpieza fueron definidas durante la etapa de validación, con el fin de garantizar que cada transformación responda a una necesidad identificada previamente y no modifique arbitrariamente la información.

---

## Flujo general de limpieza

Cada dataset posee características particulares, por lo que el proceso se desarrolla mediante tres pipelines independientes implementados en la clase `Cleaner`.

### Dataset Principal

Se realizan las siguientes transformaciones:

- Conversión de variables porcentuales almacenadas como texto a tipo numérico.
- Eliminación del símbolo `%`.
- Conversión de `POBLACIÓN_5_16` a tipo entero.
- Conversión de `CÓDIGO_ETC` al tipo entero nullable (`Int64`).
- Conversión del porcentaje de sedes conectadas a Internet a tipo `float`.

---

### Dataset PAE

Se realizan las siguientes transformaciones:

- Conversión de la columna `FECHA` al tipo `datetime`.
- Creación de la columna `AÑO`.
- Eliminación de separadores de miles en `CANTIDAD_BENEFICIARIOS_PAE`.
- Conversión de la cantidad de beneficiarios a tipo entero.

---

### Dataset ETC

Se realizan las siguientes transformaciones:

- Eliminación del registro duplicado correspondiente a la misma llave `(AÑO, CÓDIGO_ETC)`, conservando la versión consistente con el resto del conjunto de datos.
- Conversión de `POBLACIÓN_5_16` a tipo entero.
- Conversión de `TASA_MATRICULACIÓN_5_16` a tipo `float`.
- Conversión del porcentaje de sedes conectadas a Internet a tipo `float`.
- Corrección del formato decimal de `TAMAÑO_PROMEDIO_DE_GRUPO`, reemplazando comas decimales por puntos antes de su conversión numérica.

---

## Resultado esperado

Al finalizar este módulo, los tres datasets tendrán tipos de datos consistentes, formatos homogéneos y estarán preparados para la etapa de integración mediante las llaves definidas durante la validación.
