## Importación de librerías

En esta sección se importan las librerías necesarias para el notebook, así como el módulo `DataLoader`, encargado de centralizar la lectura de todos los datasets del proyecto.

```python
from pathlib import Path
import sys

# Carpeta raíz del proyecto
PROJECT_ROOT = Path.cwd().parent

# Agregar src al path de Python
sys.path.append(str(PROJECT_ROOT))

from src.data.loader import DataLoader
```

## Inicialización del DataLoader

Se crea una instancia del cargador de datos, que permitirá acceder a todos los datasets originales mediante métodos específicos.

```python
DATA_DIR = PROJECT_ROOT / "data" / "raw"

loader = DataLoader(DATA_DIR)
```

## Carga de los datasets

Se cargan los tres conjuntos de datos originales:

- **Principal:** indicadores educativos por municipio.
- **PAE:** información del Programa de Alimentación Escolar.
- **ETC:** indicadores agregados por Entidad Territorial Certificada.

Los datos permanecen sin modificaciones en esta etapa.

```python
datasets = loader.load_all()

principal = datasets["principal"]
pae = datasets["pae"]
etc = datasets["etc"]
```

## Verificación de la carga

Se inspeccionan las primeras filas de cada dataset para comprobar que la lectura fue exitosa y que las columnas fueron interpretadas correctamente.

```python
loader.summary(principal, "Dataset Principal")
```
